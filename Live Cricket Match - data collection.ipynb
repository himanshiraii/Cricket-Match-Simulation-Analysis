# *INDIA vs NEW ZEALAND, 3rd ODI MATCH, INDORE*

### What is it?

> * This python code is made to record a live cricket match.

> * This code (as of now) only records batting scorecard of a team.

> * Since I made this code to record a specific match (Ind vs NZ 3rd ODI match) that's why a few lines of codes like "print('India to chase 338.)" and playing 15 and 11 of Indian team, so you can change them as per your needs.

### How it works?

> * Firstly, you will be asked to choose your PLAYING 11 from the squad of 15, where you will enter only the jersey numbers of the players who will be playing the game.

> * After playing 11 has been selected, and since this code records batting side of a team, so next step is to select the OPENING PLAYERS of the team. (Imp-Everytime a player is choosen, his JERSEY NUMBER has to be the input and not his name)

> * And now the game begins by bowling the very first ball.

> * BALL TYPE - Now at every ball, the first thing which will be asked would be if the ball is a fair ball(f), wide(w), no ball(n) or a ball which includes just some extra runs(e). 

> * STRIKE - As soon as you ans which type of ball was bowled, the player who played that ball will be asked, so you will have to input that player's jersey number.

> * RUNS FORMAT & WICKET - Then input the runs hit by that player. Eq: write 1 if he ran a single or a 6 if he hit a six. (Imp - Numbers ranging from 0-6 are reserved for runs whereas numbers like 7,8,9 if inputted into runs scored will lead to a "wicket".)

> * WICKET - And since numbers 7/8/9 will be appended into the ball wise score, that row in the data will have to be removed later otherwise it will generate wrong results. Also, due to some last minute changes, you will get a different row for the wicket ball. (you will understand it better when applied practically). Final result will lead to 2 rows, one with runs equal to 7/8/9 and wicket=no and other row with wicket=yes. So, in post analysis you will need to drop first row (with runs=7/8/9) and keep the second one only.

> * WICKET - There is one more thing which will be asked when a wicket fells, which is "Runs along with wicket", it has been put for a case if runs are also scored along with wicket because "Runs scored will record the wicket (as designed and explained above)" and "runs along with wicket" will record runs, if nothing scored, then simpy type 0.

> * WICKET - Then input next player's jersey number who comes after the wicket falls.

> * WIDE/NOBALL - in this case after ball type is asked and "w" or "n" has been inputted, the first thing that will be asked is "w/n runs" for the extra runs that will be added when a wide/noball is bowled and I introduced this step because sometime on such balls not only a single run is added(extra) but more than 1 run situation can also happen,that's why it has been inducted in the model. So input runs keeping in mind that 1 run has to be typed in along with other runs, if any.

> * WIDE/NOBALL - Then comes a situation when more than one wide/noballs are bowled, and for that "w/n again? (y/n)" feature was put in. If yes, input y and if no, input n and follow further steps asked. 

> * WIDE/NOBALL - And later on if no more wide/noballs are not bowled after one has been done, the next ball is a legal ball hence "Runs scored" will be asked for the legal ball's score.

> * End - Input "end" in ball type when the match gets over before all the overs have been bowled. Eq: target acheived in 45 overs in a 50 over game with 5 overs left, the team gets all out before overs gets completed, etc.

# LIVE CRICKET MATCH DATA COLLECTOR

```python
import numpy as np

score={}
extras=[]
scoreboard=[]
wickets=[]


class Cricket:
    def __init__(self):
        print("IND vs NEW ZEALAND 3RD ODI,INDORE\n")
        print("India to chase 338.")
        
    # team selection   
    def team(self):       
        self.iplaying15={77:"Shubman Gill (c)",45:"Rohit Sharma",18:"Virat Kohli",1:"KL Rahul (wk)", 
                   96:"Shreyas Iyer (vc)",11:"Ayush Badoni",8:"Ravindra Jadeja",73:"Mohd Siraj",
                   22:"Harshit Rana",24:"Prasidh Krishna",23:"Kuldeep Yadav",16:"Dhruv Jurel (wk)", 
                   88:"Nitish Kumar Reddy",2:"Arshdeep Singh",64:"Yashasvi Jaiswal"}
        print(f"The Squad of of Ind for IndvsNZ 3rd ODI is: \n{self.iplaying15}")
        self.playing11=[]
        self.jersey11=[]
        for k in range(1,12):
            self.playing=int(input("\nEnter Jersey number of playerin11:"))
            if self.playing in self.iplaying15.keys():
                self.playing11.append(self.iplaying15.get(self.playing)),self.jersey11.append(self.playing)
            else:
                print("Player not found...start again")
        print(f"\nPlaying 11 of Ind vs NZ 3rd ODI match,Indore is: \n{self.playing11}")
        print(f"\n Corresponding Jersey numbers of players: \n{self.jersey11}")
        
        
    # wickets procedure
    def wicket(self):
        self.outplayer=int(input("Enter out player jersey no.:"))
        self.outway=input("Enter way of out:")
        self.ways=["runout","bowled","caught","lbw","stumped","hitwicket","timeout","obsfield","mankad"]
        if self.outway in self.ways:
            if self.outplayer==self.player1:
                print(f"Jersey {self.outplayer}-{self.iplaying15.get(self.outplayer)} lost their wicket.")
                self.player1=int(input("Next player 1:"))
            elif self.outplayer==self.player2:
                print(f"Jersey {self.outplayer}-{self.iplaying15.get(self.outplayer)} lost their wicket.")
                self.player2=int(input("Next player 2:"))
            scoreboard.append((self.iplaying15.get(self.outplayer),self.run2,"Yes",self.outway,f"{self.h}.{self.k}"))
            wickets.append(f"{self.iplaying15.get(self.outplayer)}({self.outway})")
                
                
    # format for a fair ball to be bowled     
    def runsformat(self):
        self.striker=int(input("Who's on strike?(jersey no.):"))
        self.run=int(input("Runs scored:"))
        if self.run in [0,1,2,3,4,5,6]:
            score[f"{self.h}.{self.k}"]=self.run
            self.strike()
        else:
            self.run2=int(input("Runs along with wicket?:"))
            score[f"{self.h}.{self.k}"]=["W",self.run2]
            self.strike()
            self.wicket()
        
            
    # format when wide/no-balls are bowled        
    def combo(self):
        self.r=int(input(f"{self.btype} runs:"))             # no condition on limit of runs as of other possible ways of getting more runs on that wide ball and noy just a single
        score[f"{self.h}.{self.k}.{self.btype}"]=self.r      # adding nb score to the scoreboard
        self.m=scoreboard.append(("Extras",self.r,"No","No",f"{self.h}.{self.k}.{self.btype}"))
        extras.append(self.r)                                # extra runs added separately
        self.y=0
        while True:
            self.y=self.y+1
            self.balltype=input(f"{self.btype} again?(y/n):")
            if self.balltype=="y":
                self.rr=int(input(f"More {self.btype} runs:"))
                score[f"{self.h}.{self.k}.{self.btype}.{self.y}"]=self.rr
                self.w=scoreboard.append(("Extras",self.rr,"No","No",f"{self.h}.{self.k}.{self.btype}.{self.y}"))
                extras.append(self.rr)
            else:
                break
        self.runsformat()
        

    # function made for both wide/noball which is connected to "def play" from where the whole process follows                           
    def wideandnoball(self):
        if self.btype=="w":
            self.combo()
        elif self.btype=="n":
            self.combo()
        else:
            print("Incorrect input")
            
         
    # for storing wides/noballs data                           
    def wides(self):
        if self.btype=="w" or self.btype=="n":
            scoreboard.pop()
            self.m
            if self.balltype=="y":
                self.w
            else:
                scoreboard.append((self.c,self.run,"No","No",f"{self.h}.{self.k}"))
       
     
    # data to be recorded based on which player was on strike    
    def strike(self):
        if self.striker==self.player1:
            self.n=self.iplaying15.get(self.striker)
            scoreboard.append((self.n,self.run,"No","No",f"{self.h}.{self.k}"))
            self.c=self.n
            self.wides()
        elif self.striker==self.player2:
            self.n1=self.iplaying15.get(self.striker)
            scoreboard.append((self.n1,self.run,"No","No",f"{self.h}.{self.k}"))
            self.c=self.n1
            self.wides()
        else:
            print("No such player found.")
            self.runsformat()
    
            
    # criteria for extra runs of leg bye    
    def extraruns(self):
        self.e=int(input("Extra runs of legbye/bye:"))   
        scoreboard.append(("Extras",self.e,"No","No",f"{self.h}.{self.k}.lb"))                      
        score[f"{self.h}.{self.k}.extra"]=self.e
        extras.append(self.e)
        self.ifout=input("Wicket?(y/n):")
        if self.ifout=="y":
            scoreboard.pop()
            self.run2=self.e
            self.wicket()
            scoreboard.append(("Extras",self.run2,"Yes",self.outway,f"{self.h}.{self.k}"))                      
            
        
    # encapsulated function for whole game    
    def play(self):
        self.team()
        self.player1=int(input("\nEnter player 1 on crease:"))
        self.player2=int(input("Enter player 2 on crease:"))
        if (self.player1 and self.player2 in self.iplaying15) and (self.player1!=self.player2):
            for self.h in range(0,50):            # overs -- change the range (50 for ODI, 20 for T20 or how many you want)
                for self.k in range(1,7):         # balls
                    self.btype=input(f"\n{self.h}.{self.k} \nfair(f)/wide(w)/noball(n)/extra(e)/end:")
                    if self.btype=="f":    # fair delivery
                        self.runsformat()
                        print()
                    elif self.btype=="w" or self.btype=="n":   # wide or no ball
                        self.wideandnoball()
                        print()
                    elif self.btype=="end":    # end the game (in case when overs are left but game has ended)--over
                        break
                    elif self.btype=="e":     # other extra runs
                        self.extraruns() 
                    else:
                        print("Incorrect Input")
                if self.btype=="end":     # end the game (in case when overs are left but game has ended)--whole match
                    break
        else:
            print("Unknown/Same players found.")
```
